```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**Matplotlib Basics**

```python
x = np.linspace(0,5,11)
x
y = x **3
y
```

```python
plt.plot(x,y)
plt.title("Hello")
plt.xlabel("x axis")
plt.ylabel("y axis")
```

**Subplots in Matplotlib**

```python
plt.subplot(2,2,1)
plt.plot(y,x)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(x,x)
plt.subplot(2,2,3)
plt.plot(x,x**y)
```

**Object Oriented way**

```python
fig = plt.figure(figsize=(8,8), dpi= 100)
axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.4,0.4,0.5])
axis2.plot(x,y)
```

**Types of Plots**

```python
plt.scatter(x,y)
```

```python
from random import sample 
data = sample(range(1,1000),100)
plt.hist(data)
```

```python
plt.boxplot(x)
```

**Saving Plots**

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
plt.savefig("basicplot.png")
```
